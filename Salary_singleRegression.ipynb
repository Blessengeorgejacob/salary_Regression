```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
font1={'family':'serif','color':'red','size':30}
font2={'family':'serif','color':'blue','size':20}
plt.xlabel("exp",fontdict=font2)
plt.ylabel("sal",fontdict=font2)
plt.title("exp-sal",fontdict=font1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred)
```

```python
df1=pd.DataFrame({"Actual_value":y_test,"predicted_value":y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(y_test,y_pred)
error
```

```python
from sklearn.metrics import mean_absolute_percentage_error
per=mean_absolute_percentage_error(y_test,y_pred)
per
```

```python
from sklearn.metrics import mean_squared_error
squ=mean_squared_error(y_test,y_pred)
squ
```

```python
from sklearn.metrics import mean_squared_error
sq=np.sqrt(squ)
sq
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
